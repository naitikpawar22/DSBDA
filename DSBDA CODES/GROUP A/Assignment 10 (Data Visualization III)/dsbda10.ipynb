```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
import matplotlib.pyplot as plt

iris_df.hist()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(x = Iris['PetalWidthCm'], kde=True)
plt.title('Distribution of Petal Width')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplots for each feature separately
sns.boxplot(x = Iris['SepalLengthCm'])
plt.title('Boxplot of Sepal Length')
plt.show()

sns.boxplot(x = Iris['SepalWidthCm'])
plt.title('Boxplot of Sepal Width')
plt.show()

sns.boxplot(x = Iris['PetalLengthCm'])
plt.title('Boxplot of Petal Length')
plt.show()

sns.boxplot(x = Iris['PetalWidthCm'])
plt.title('Boxplot of Petal Width')
plt.show()

# Boxplot comparing feature vs species
sns.boxplot(x = 'Species', y = 'SepalLengthCm', data = Iris)
plt.title('Sepal Length vs Species')
plt.show()

sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = Iris)
plt.title('Petal Length vs Species')
plt.show()
```
